# Predictive Model looks pretty good now let's build the Confusion Matrix

```python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(pap_data0609.Sold, predictions)

print(cnf_matrix)

print("Accuracy:",metrics.accuracy_score(pap_data0609.Sold, predictions))

print("Precision:",metrics.precision_score(pap_data0609.Sold, predictions))

print("Recall:",metrics.recall_score(pap_data0609.Sold, predictions))
```
