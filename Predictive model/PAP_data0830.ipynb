# Data cleaning and transformation

```python
title_list=[i.strip().replace('\xa0', '').replace('\t', ' ').lstrip('Location') for i in locations_data.Title.tolist()]
```

```python
import re
price_list=[]
for i in title_list:
    try:
        match=re.search(r'(\d?\.?\d+)€', i).group(1)
    except:
        match='-1'
    price_list.append(match)

price_list1=[int(i.replace('.','')) for i in price_list]
    
price_list1
```

```python
title_list
```

```python
location=[]
for i in title_list:
    try:
        match=re.search(r'Paris [\d+\w+]*', i).group(0)
    except:
        match='-1'
    location.append(match)
location
```

```python
m2=[]
for i in title_list:
    try:
        match=re.search(r'\d+m', i).group(0)
    except:
        match='-1'
    m2.append(match)
m2
```

```python
description=[title_list[i].replace(price_list[i],'').strip('€ ').replace(location[i],'').replace(m2[i],'').replace(' ² ',' 1').replace('é', 'e') for i in range(len(title_list))]
loc_type=[]
for i in description:
    try:
        match=re.search(r'^\s*\D*', i).group(0)
    except:
        match='-1'
    loc_type.append(match)
loc_type
```

```python
date=locations_data.Date.tolist()
date
```

```python
import re
loc_date=[]
for i in date:
    try:
        match=re.findall(r"\/  \d\d\D*\d*", str(i))
    except:
        match='-1'
    loc_date.append(match)
#loc_date = ([str(i).strip('/  ') for i in loc_date])
loc_date
```
